# Logistic Regression model

```python
# import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
# to ignore warning messages
import warnings
warnings.filterwarnings('ignore')
```

```python
# no.of columns
len(df.columns)
```

```python
# effected people are 1 and non effected are 0
df['Diagnosis (ALS)'].value_counts()
```

```python
# fix features and target
X = df.drop(columns='Diagnosis (ALS)') # features
y = df['Diagnosis (ALS)'] # target
```

```python
# train the model
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=20, stratify=y)
```

```python
# define the model 
model = LogisticRegression()
```

```python
# fit the model
model.fit(x_train,y_train)
```

```python
# predict with model
pred = model.predict(x_test)
accuracy_score(y_test,pred)
```

```python
#confusion matrix
confusion_matrix(y_test,pred)
```

```python
# classification report
class_report= classification_report(y_test,pred)
```

```python
print(class_report)
```

# cross validation

```python
# import necessary libraries
from sklearn.model_selection import cross_val_score,KFold,LeaveOneOut
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
# create val and models
val = [KFold(8),LeaveOneOut()]
models = [LogisticRegression(multi_class='multinomial', solver='lbfgs', random_state=42),
          DecisionTreeClassifier(random_state=42),
          GaussianNB(),
          RandomForestClassifier(n_estimators=100, random_state=42),
          KNeighborsClassifier(n_neighbors=5),
          SVC(kernel='linear', random_state=42),
         ]
```

```python
for i in val:
    for j in models:
        score  = cross_val_score(estimator=j,cv=i,X=X,y=y)
        print(j)
        print(i)
        print(score.mean())
        print('----'*10)
```

# Grid search

```python
# import necessary libraries
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
# Define models
gnb_model = GaussianNB()
dt_model = DecisionTreeClassifier()
rf_model = RandomForestClassifier()
```

```python
# Define hyperparameter grids

gnb_param_grid = {
    'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]
}

dt_param_grid = {
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

rf_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
# List of models and their parameter grids
models = [
    ('GaussianNB', gnb_model, gnb_param_grid),
    ('Decision Tree', dt_model, dt_param_grid),
    ('Random Forest', rf_model, rf_param_grid)
]
```

```python
# Initialize GridSearchCV
for name, model, param_grid in models:
    print(f"Tuning hyperparameters for {name}...")
grid_search = GridSearchCV(estimator=model , param_grid=param_grid, cv=5,scoring='accuracy', n_jobs = -1 , verbose=2)
grid_search.fit(x_train, y_train)
```
