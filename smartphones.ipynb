# linear regression

```python
# import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.feature_selection import SelectKBest,chi2,f_classif
```

```python
# fix feature and target columns
X = data.drop(columns='price')
y = data['price']
```

```python
# Encode categorical variables if any
X = pd.get_dummies(X, drop_first=True)
```

```python
X
```

```python
# to select best 8 columns
kbest8 = SelectKBest(chi2,k=8)
best8 = kbest8.fit(X,y)
```

```python
X.columns[best8.get_support()]
```

```python
# load the new data
X = data[['battery_capacity', 'internal_memory', 'refresh_rate',
       'primary_camera_rear', 'primary_camera_front', 'os',
       'has_5g', 'ram_capacity']]
```

```python
# Apply log transformation to the target y
y_log = np.log(y)
```

```python
# Splitting data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_log, test_size=0.2, random_state=42)
```

```python
# Creating the model
model = LinearRegression()
```

```python
# Training the model
model.fit(X_train, y_train)
```

```python
# Making predictions
y_pred = model.predict(X_test)
```

```python
# Evaluating the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
# prediction price of new smartphone
new_smartphone = [[5000, 128, 120, 64, 32, 0, 1, 8]]  # Example features (replace with actual values)
```

```python
predicted_price = model.predict(new_smartphone)
print(predicted_price)
```

```python
# convert predicted price to normal price
og = np.exp(predicted_price)
print(og)
```
